# Keras API 다루기


Keras에서 모델을 정의할 때, 두 가지 주요 API가 있습니다: Sequential API와 Functional API입니다. 두 API는 모두 Keras 모델을 구성할 수 있지만, 사용 방법과 기능에 차이가 있습니다. 아래에서 이 두 API의 주요 차이점을 설명합니다.

### 1. Sequential API

**Sequential API**는 모델을 층별로 순차적으로 쌓아 올리는 방식입니다. 직관적이고 사용하기 간편하여 간단한 모델을 구축할 때 유용합니다.

#### 특징:
- **단순성**: 모델을 층별로 순차적으로 쌓는 방식으로 코드가 간단하고 직관적입니다.
- **선형 구조**: 모델의 구조가 직선형(즉, 입력층 → 은닉층 → 출력층)으로 되어 있어 복잡한 연결이 필요한 경우에는 제한적입니다.

#### 예제:
```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(input_dim,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

- `Sequential` 모델에 `Dense` 층을 차례로 추가하여 모델을 구성합니다.
- `input_shape`는 첫 번째 층에서만 지정하며, 이후 층은 자동으로 이전 층의 출력 크기를 입력으로 받습니다.

### 2. Functional API

**Functional API**는 더 복잡한 모델을 정의할 수 있는 방법입니다. 모델의 구조가 비선형적이거나, 다중 입력/출력, 잔차 연결 등 복잡한 모델을 구성할 수 있습니다.

#### 특징:
- **유연성**: 다양한 형태의 모델(예: 다중 입력/출력, 분기, 병합 등)을 구현할 수 있습니다.
- **복잡한 모델**: 비선형적 또는 복잡한 네트워크 구조를 정의할 때 유용합니다.

#### 예제:
```python
from keras.models import Model
from keras.layers import Input, Dense

inputs = Input(shape=(input_dim,))
x = Dense(64, activation='relu')(inputs)
x = Dense(64, activation='relu')(x)
outputs = Dense(1, activation='sigmoid')(x)

model = Model(inputs=inputs, outputs=outputs)
```

- `Input` 층을 정의하여 모델의 입력을 명시합니다.
- 층들을 함수 호출처럼 연결하여 네트워크를 구성합니다.
- `Model` 객체를 생성할 때 `inputs`와 `outputs`을 명시하여 전체 모델을 정의합니다.

### 비교

| 특징              | Sequential API                                   | Functional API                                |
|-------------------|--------------------------------------------------|----------------------------------------------|
| **구조**           | 선형적 구조 (입력 → 은닉층 → 출력)                    | 비선형적 구조 (분기, 병합 등 복잡한 구조 가능)        |
| **사용 용이성**    | 간단하고 직관적                                    | 복잡한 구조를 지원하지만 코드가 더 복잡할 수 있음 |
| **모델 복잡성**    | 간단한 모델에 적합                                   | 복잡한 모델에 적합                            |
| **입력/출력**      | 단일 입력/출력                                     | 다중 입력/출력 가능                           |

### 결론
- **Sequential API**는 빠르고 간편하게 모델을 만들 수 있는 방법으로, 단순한 순차적 모델에 적합합니다.
- **Functional API**는 복잡한 모델을 구성할 수 있는 유연성을 제공하며, 다양한 구조의 네트워크를 정의할 때 유리합니다.

모델의 복잡성에 따라 적절한 API를 선택하면 됩니다.

MNIST DATASET

MNIST 데이터셋은 머신러닝 및 딥러닝에서 가장 널리 사용되는 데이터셋 중 하나로, 주로 손글씨 숫자 인식 문제를 다루는 데 사용됩니다. "MNIST"는 "Modified National Institute of Standards and Technology"의 약자로, 원래의 NIST 데이터셋을 기반으로 한 변형입니다.

### MNIST 데이터셋의 구성

1. **데이터셋 구성:**
   - **이미지**: 28x28 픽셀의 흑백 이미지입니다. 각 이미지는 손글씨로 쓴 숫자(0부터 9까지)를 나타냅니다.
   - **레이블**: 각 이미지에 해당하는 숫자 레이블입니다.

2. **데이터 분할:**
   - **훈련 세트 (Training Set)**: 60,000개의 이미지와 해당 레이블로 구성됩니다. 모델을 학습시키는 데 사용됩니다.
   - **테스트 세트 (Test Set)**: 10,000개의 이미지와 해당 레이블로 구성됩니다. 모델의 성능을 평가하는 데 사용됩니다.

3. **데이터 형식:**
   - 각 이미지는 28x28 픽셀의 그레이스케일 이미지로, 픽셀 값은 0부터 255까지의 범위를 가지며, 0은 완전히 검정색, 255는 완전히 흰색을 의미합니다.
   - 이미지는 1차원 배열로 펼쳐져 있어, 각 이미지가 784(28x28)개의 픽셀 값으로 구성됩니다.

### MNIST 데이터셋의 특징

- **단순성과 기본성**: MNIST 데이터셋은 머신러닝 및 딥러닝의 기본적인 알고리즘을 테스트하고 이해하는 데 적합합니다. 구조가 간단하여 알고리즘의 성능을 쉽게 평가할 수 있습니다.
- **표준 데이터셋**: 다양한 모델과 알고리즘을 비교하는 데 널리 사용되는 표준 데이터셋입니다.

### 데이터셋의 역사와 영향

- **역사**: MNIST는 1998년 Yann LeCun, Corinna Cortes, and Christopher J.C. Burges가 발표한 데이터셋으로, 기계 학습 분야의 초기 연구와 실험에서 중요한 역할을 했습니다.
- **영향**: MNIST는 딥러닝 알고리즘의 성능을 평가하는 데 사용되는 데이터셋으로, 많은 연구와 개발에서 기준으로 사용되었습니다. 특히, CNN(Convolutional Neural Network)과 같은 모델의 발전에 중요한 기여를 했습니다.

이 데이터셋은 머신러닝 입문자부터 연구자까지 폭넓게 사용되며, 기본적인 알고리즘과 모델을 학습하고 평가하는 데 유용한 자료입니다.

---

````python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.utils import plot_model
````

````python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
````

````python
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
````

````python
model = Sequential()
model.add(Flatten(input_shape=(28, 28)))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))
````

````python
model.compile(optimizer = 'adam',
              loss = 'categorical_crossentropy',
              metrics = ['accuracy'])
````

````python
model.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
````

````python
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
````

````python
y_pred = model.predict(x_test)
````

````python
y_pred
````

### Fashion MNIST 데이터셋의 구성
**Fashion MNIST** 데이터셋은 원래 MNIST 데이터셋의 손글씨 숫자 이미지를 패션 아이템으로 대체한 데이터셋입니다.   
Fashion MNIST는 동일한 이미지 크기와 형식(28x28 픽셀의 그레이스케일 이미지)을 사용하지만, 각 이미지는 다양한 패션 아이템을 나타냅니다.  

1. **데이터셋 구성:**
   - **이미지**: 28x28 픽셀의 흑백 이미지입니다. 각 이미지는 패션 아이템(예: 티셔츠, 바지, 신발 등)을 나타냅니다.  
   - **레이블**: 각 이미지에 해당하는 패션 아이템의 레이블입니다.   
   총 10개의 클래스가 있습니다.  

2. **데이터 분할:**
   - **훈련 세트 (Training Set)**: 60,000개의 이미지와 해당 레이블로 구성됩니다.
   - **테스트 세트 (Test Set)**: 10,000개의 이미지와 해당 레이블로 구성됩니다.

3. **클래스 레이블:**
   - 0: T-shirt/top
   - 1: Trouser
   - 2: Pullover
   - 3: Dress
   - 4: Coat
   - 5: Sandal
   - 6: Shirt
   - 7: Sneaker
   - 8: Bag
   - 9: Ankle boot

### Fashion MNIST 데이터셋의 특징

- **패션 아이템**: MNIST 데이터셋과 비슷한 이미지 형식을 사용하지만, 실제 패션 아이템을 인식하는 문제를 해결하도록 설계되었습니다.
- **다양성**: 패션 아이템의 다양한 형태와 디자인으로 인해 모델의 학습과 평가에서 더 많은 변화를 제공합니다.

### 데이터셋의 역사와 영향

- **역사**: Fashion MNIST는 2017년 Zalando Research에서 발표되었습니다.   
원래 MNIST 데이터셋의 한계를 보완하기 위해, 실제 패션 아이템 이미지를 사용하는 더 어려운 문제를 제공하기 위해 개발되었습니다.
- **영향**: Fashion MNIST는 MNIST와 비슷하지만 더 복잡한 문제를 제공하며, 딥러닝 알고리즘의 성능을 평가하는 데 더 유용합니다.  
 특히, 이미지 분류와 관련된 연구에서 유용하게 사용됩니다.  

### 결론

Fashion MNIST는 패션 아이템 이미지를 사용하여 모델의 성능을 평가하는 데이터셋으로, MNIST와 유사한 형식을 가지고 있지만 더 복잡한 패션 아이템 이미지를 다루므로 모델의 성능을 더 세밀하게 평가할 수 있습니다. 이는 기본적인 이미지 분류 문제를 연구하거나 테스트할 때 유용한 데이터셋입니다.

1. **데이터 출처**:
   - 이 데이터셋은 1990년 미국 인구 조사 데이터를 기반으로 하며, 캘리포니아 주택 및 지역 개발 부서에서 제공하였습니다. Scikit-Learn의 `sklearn.datasets` 모듈을 통해 사용할 수 있습니다.

2. **특징 (Features)**:
   - 데이터셋에는 다음과 같은 주택 관련 정보가 포함되어 있습니다:
     - `MedInc` (Median Income): 블록 내 가구의 중간 소득.
     - `HouseAge` (House Age): 블록 내 주택의 중간 연령.
     - `AveRooms` (Average Rooms): 가구당 평균 방 개수.
     - `AveOccup` (Average Occupancy): 가구당 평균 인원 수.
     - `Latitude`: 블록 그룹의 위도.
     - `Longitude`: 블록 그룹의 경도.

3. **타겟 (Target)**:
   - `MedHouseVal` (Median House Value): 블록 그룹의 주택 중간 가격 (단위: 천 달러). 이 값이 모델의 예측 목표입니다.

4. **데이터 분할**:
   - 데이터셋은 일반적으로 훈련 세트와 테스트 세트로 나누어 모델 학습과 평가에 사용됩니다.

````python
from sklearn.model_selection import train_test_split

X_train_full, X_test, y_train_full, y_test = train_test_split(
    housing.data, housing.target, random_state=42)
X_train, X_valid, y_train, y_valid = train_test_split(
    X_train_full, y_train_full, random_state=42)
````

````python
norm_layer = tf.keras.layers.Normalization(input_shape = X_train.shape[1:])
````

````python
model = tf.keras.Sequential([
    norm_layer,
    tf.keras.layers.Dense(30, activation='relu'),
    tf.keras.layers.Dense(30, activation='relu'),
    tf.keras.layers.Dense(30, activation='relu'),
    tf.keras.layers.Dense(30, activation='relu'),
    tf.keras.layers.Dense(1)
])
````

----

 # categorical_crossentropy, binary_crossentropy, 그리고 sparse_categorical_crossentropy

 특히 분류 문제에서는 모델이 얼마나 잘 예측하는지를 평가하는 데 사용 `categorical_crossentropy`, `binary_crossentropy`, 그리고 `sparse_categorical_crossentropy`는 각각 다른 유형의 분류 문제에 적합한 손실 함수입니다.

### 1. Categorical Crossentropy (`categorical_crossentropy`)

**Categorical Crossentropy**는 다중 클래스 분류 문제에서 사용하는 손실 함수입니다. 이 손실 함수는 각 클래스에 대해 확률을 출력하는 모델에서 사용됩니다. 특히, 모델의 출력은 소프트맥스 활성화 함수로 확률 분포로 변환됩니다.

#### 특징:
- **용도**: 다중 클래스 분류 문제에서, 클래스 레이블이 원-핫 인코딩(각 클래스에 대해 이진 벡터)으로 표현될 때 사용됩니다.
- **입력**: 모델의 출력은 각 클래스에 대한 확률값을 포함하는 벡터입니다. 타겟 레이블은 원-핫 인코딩된 벡터입니다.

#### 예제:
- 클래스가 3개인 경우:
  - 모델 출력: `[0.1, 0.7, 0.2]` (각 클래스에 대한 확률)
  - 실제 레이블: `[0, 1, 0]` (정답 클래스는 두 번째 클래스)
  
- 손실 함수는 모델이 출력한 확률 분포와 실제 원-핫 인코딩된 레이블 간의 차이를 측정합니다.

### 2. Binary Crossentropy (`binary_crossentropy`)

**Binary Crossentropy**는 이진 분류 문제에서 사용하는 손실 함수입니다. 이 손실 함수는 모델의 출력이 두 가지 클래스 중 하나일 확률을 출력할 때 사용됩니다. 모델의 출력은 시그모이드 활성화 함수로 변환되어 0과 1 사이의 확률값을 가집니다.

#### 특징:
- **용도**: 이진 분류 문제에서, 출력이 0과 1 사이의 값으로 표현될 때 사용됩니다.
- **입력**: 모델의 출력은 0과 1 사이의 확률값이며, 타겟 레이블은 0 또는 1입니다.

#### 예제:
- 이진 분류의 경우:
  - 모델 출력: `0.8` (클래스 1의 확률)
  - 실제 레이블: `1` (정답 클래스는 클래스 1)

- 손실 함수는 모델의 예측 확률과 실제 이진 레이블 간의 차이를 측정합니다.

### 3. Sparse Categorical Crossentropy (`sparse_categorical_crossentropy`)

**Sparse Categorical Crossentropy**는 다중 클래스 분류 문제에서 사용되지만, `categorical_crossentropy`와는 달리 타겟 레이블이 원-핫 인코딩이 아닌 정수 인덱스 형태로 제공됩니다.

#### 특징:
- **용도**: 다중 클래스 분류 문제에서, 타겟 레이블이 원-핫 인코딩 대신 클래스 인덱스(정수형)으로 제공될 때 사용됩니다.
- **입력**: 모델의 출력은 각 클래스에 대한 확률값을 포함하는 벡터입니다. 타겟 레이블은 정수형 클래스 인덱스입니다.

#### 예제:
- 클래스가 3개인 경우:
  - 모델 출력: `[0.1, 0.7, 0.2]` (각 클래스에 대한 확률)
  - 실제 레이블: `1` (정답 클래스는 두 번째 클래스)

- 손실 함수는 모델이 출력한 확률 분포와 실제 정수형 클래스 레이블 간의 차이를 측정합니다.

### 요약

| 손실 함수                   | 사용 용도                                      | 타겟 레이블 형태               | 출력 레이어 활성화 함수  |
|-----------------------------|------------------------------------------------|-------------------------------|-------------------------|
| **Categorical Crossentropy** | 다중 클래스 분류 문제, 원-핫 인코딩된 레이블 | 원-핫 인코딩된 벡터            | 소프트맥스                |
| **Binary Crossentropy**      | 이진 분류 문제, 0 또는 1로 표현된 레이블     | 0 또는 1                        | 시그모이드                |
| **Sparse Categorical Crossentropy** | 다중 클래스 분류 문제, 정수형 레이블         | 정수형 클래스 인덱스            | 소프트맥스                |

이 손실 함수들은 각각의 문제 유형과 데이터 표현 방식에 맞게 모델을 학습시키기 위한 적절한 선택입니다.
